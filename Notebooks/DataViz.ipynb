```python
# Ahora visualizaremos los datos por comarcas y municipios

def comarcas(regepa):
    regepa_comarca = regepa.groupby(['COMARCA']).mean().reset_index()

    rgp = regepa_comarca.sort_values('SUPERFICIE CULTIVADA (hect√°reas)', ascending=False).reset_index()

    return rgp

def municipios(regepa):
    regepa1 = regepa.groupby(['MUNICIPIO']).sum()

    rgp = regepa1.sort_values('SUPERFICIE CULTIVADA (hect√°reas)', ascending=False).reset_index()

    return rgp
```
